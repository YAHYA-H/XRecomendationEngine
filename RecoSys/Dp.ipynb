```python
# Create a smaller dataframe with a subset of all features
small_df = df[['title', 'genres', 'budget', 'revenue', 'runtime', 'release_date']]

# output only the first five rows of small_df
small_df.head()
```

```python
# Get information of the data types of each features
small_df.info()
```

```python
import numpy as np

# convert budget to float
# df['budget'].astype('float') #fails-valueError cos of field has a str

# Func to convert to float manually
def to_float(x):
    try:
        x = float(x)
    except:
        x = np.nan
    return x

# Apply the to_float func to all values in the budget col
small_df['budget'] = small_df['budget'].apply(to_float) #apply() allow us to apply a func to every field in particular col

#try converting to float using pandas astype
small_df['budget'] = small_df['budget'].astype('float')

#Get the data types for all features
small_df.info()
```

```python
#convert release_date into pandas datetime format
small_df['release_date'] = pd.to_datetime(small_df['release_date'], errors='coerce')

#Extracts year from the datetime
small_df['year'] = small_df['release_date'].apply(lambda x: str(x).split('-')[0]
                                                  if x !=np.nan else np.nan)

#Display the Dateframe with the new 'year' features
small_df.head()


                                                  
```

```python
# sort dataframee based on release year
small_df = small_df.sort_values('year')
small_df.head()
```

```python
# sort Movies based on revenue(in descending order)
small_df = small_df.sort_values('revenue', ascending=False)
small_df.head()
```

```python
# select only movies which earned more than 1 billion
new = small_df[small_df['revenue'] > 1e9]
new
```

```python
# select only those movies which earned more than 1billion and spent less than 150million

new2 = small_df[(small_df['revenue'] > 1e9) & (small_df['budget'] > 1.5e8) ]
new2
```
